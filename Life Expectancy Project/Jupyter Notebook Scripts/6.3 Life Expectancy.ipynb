```python
# Create a subplot with `fig` and `ax`.
fig, ax = plt.subplots(1, figsize=(15, 15))
```

```python
plt.title('Life Expectancy by Country', size=30)
tl = world_clean2.plot(column='Life expectancy ',
cmap='Reds',      # Color map for the states                     
             linewidth=0.6,      # Line width for state borders
             ax=ax,              # Plotting the map on `ax`
             edgecolor='black'); # State border colors
```

```python
c_bar_colors = plt.cm.ScalarMappable(cmap='Reds', norm=plt.Normalize(vmin=world_clean2['Life expectancy '].min(), vmax=world_clean2['Life expectancy '].max()))

# Plot the colorbar.
color_bar = fig.colorbar(c_bar_colors)
```
